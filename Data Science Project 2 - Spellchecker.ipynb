```python
import nltk 
from nltk.metrics import edit_distance
from nltk.corpus import words
from nltk.book import *
```

I had always been fascinated how a computer could predict and correct your spelling. When we were asked to study nltk for our project it became clear that I could implement some of these tools to create my own spellchecker.

Firstly, I had to decide which corpus would be suitable to use for this task. While searching through the various corpora available in nltk I came across the word corpus.This corpus consists of 250,000 words which are all correctly spelt. This corpus is already used by existing spellcheckers so it was definitely a good corpus to use for this project.

```python
correct_words = nltk.corpus.words.words()
```

```python
def test(bad_word):
    sorted_list = [w for w in correct_words if (len(w)-len(bad_word))**2<=4]
    #sorted_list will be a list which contains all the words that are longer than the 'bad_word' by 2 and also the ones shorter by 2
    closest_so_far = 1000 
    shortlist = []
    for w in sorted_list:
        current_distance = edit_distance(bad_word,w)
        if current_distance==closest_so_far:
            shortlist.append(w)
        elif current_distance < closest_so_far:
            closest_so_far = current_distance
            shortlist=[w]
    return shortlist
```

```python
test('theg')
```

I started out with the code above, it worked fine but it was extremely slow. This was because it was filtering through the huge word file and creating sublists each time it was being run. I realised this run time was not viable so I created a dictionary comprehension which created these sublists before the code was run. This greately improved the speed as it was no longer filtering through the huge word file each time a new word was entered only at the start.

```python
filtered_lists = {i: [w for w in correct_words if (len(w)-i)**2<=4] for i in range(1,10)}
```

Finally, I combined my various parts of code under one function called spellcheck seen below.

```python
def spellcheck(incorrect_word):

    closest_so_far = 1000
    shortlist = []
    for word in filtered_lists[len(incorrect_word)]:
        current_distance = edit_distance(incorrect_word,word)
        if current_distance==closest_so_far:
            shortlist.append(word)
        elif current_distance < closest_so_far:
            closest_so_far = current_distance
            shortlist=[word]

    frequency = FreqDist(w for w in text1 if w in shortlist)
    most_common = [word for (word,freq) in frequency.most_common()]
    if len(shortlist)==1:
        print('Best option:',shortlist[0])
    else:
        print('Best option:',most_common[0],'\nOther options:',most_common[1:])
        
```

```python
spellcheck('theg')
```
